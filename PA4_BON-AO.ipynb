#### For part a.

##### Filename: Instru = [“Name”, “GEAS”, “Electronics >70”]; where track is constant as
##### Instrumentation and hometown Luzon


```python
Instru = df_board.loc[(df_board['Track'] == 'Instrumentation') # Finds tracks that have instrumentation
            & (df_board['Hometown'] == 'Luzon') # Locates info wherein their hometown is in Luzon
            & (df_board['Electronics'] > 70)][['Name', 'GEAS', 'Electronics']] 
            # Finds all who passed electronics  # Gets their name, GEAS, and their grade in elecs

#Finally stores them to the new data frame named Instru

Instru
```

#### For part b.

##### Mindy = [ “Name”, “Track”, “Electronics”, “Average >=55”]; where hometown is
##### constant as Mindanao and gender Female


##### To first get the required dataframe for Mindy, we need to find first the averages of all students as it is not given in the downloaded excel file.

```python
df_board['Average'] = df_board[['Math', 'GEAS', 'Electronics']].mean(axis=1) # gets the mean of the 4 scored parts of the board
# creates a new column labeled as 'Average' in the data frame

df_board
```

##### We are now able to proceed to create the dataframe named Mindy as we have now created a new column
##### for all the takers that shows their average grade between the four parts of the board exam

```python
Mindy = df_board[(df_board['Hometown'] == 'Mindanao') # Condition to find students whose hometown is in Mindanao
        & (df_board['Gender'] == 'Female') # Condition to find all female takers
        & (df_board['Average'] >= 55)] [['Name', 'Track', 'Electronics', 'Average']] ##Condition for their average must be equal or higher than 55    
            #Gets Info of their Name, Track, Elec grade, and average grade                

#Finally stores the new sliced and subsetted dataframe to 'Mindy'

Mindy
```

#### 2. Create a visualization that shows how the different features contributes to average grade. Does
#### chosen track in college, gender, or hometown contributes to a higher average score?

##### First, we will check if the chosen track in college affects the averafe grade of the board takers
##### We can do this by using a boxplot to visualize if it has any effect at all or not.

```python
#sb to call the library of seaborn

sb.boxplot(x='Track', y='Average', data=df_board, palette='Set3').set_title('Average Score by Track') 

#Read left to right

# Treats the tracks as the x-axis while the average grade is the y-axis

# Data used is from df_board

# Use pallete as a colorway for the visualized data

# Set title as 'Average Score by Track'

sb.despine() # Removes top line for a cleaner look
```

##### Secondly, we will check if the gender of the taker affects their average grade in the board exam
##### We can do this by using a violinplot to visualize if it has any effect at all or not.

```python
#sb to call the library of seaborn

sb.violinplot(x='Gender', y='Average', data=df_board, palette='Set2').set_title('Average Score by Gender')

# Read left to right

# Treats the genders as the x-axis while the average grade is the y-axis

# Data used is from df_board

# Use pallete as a colorway for the visualized data

# Set title as 'Average Score by Gender'

sb.despine() # Removes top line for a cleaner look
```

##### Lastly, to show that if the hometown of the board taker affects their average grade
##### We can visualize this by using a bar graph 

```python
#sb to call the library of seaborn

sb.barplot(x='Hometown', y='Average', data=df_board, palette='Set1').set_title('Average Score by Hometown')

# Read left to right

# Treats the hometowns as the x-axis while the average grade is the y-axis

# Data used is from df_board

# Use pallete as a colorway for the visualized data

# Set title as 'Average Score by Hometown'

sb.despine() # Removes top line for a cleaner look
```
